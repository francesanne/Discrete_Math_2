**1. To prove that $e$ is irrational**

we'll start by assuming the opposite, that $e$ is rational, meaning it can be expressed as a ratio of two integers $\frac{a}{b}$, where $a$ and $b$ are integers with no common factors other than 1.

Now, let's recall the series expansion of $e$:

$$ e = \sum_{k=0}^{\infty} \frac{1}{k!} = 1 + \frac{1}{1!} + \frac{1}{2!} + \frac{1}{3!} + \cdots $$

Each term in this sum is positive, so if $e = \frac{a}{b}$, then $\frac{a}{b} > 0$, which means $a > 0$ and $b > 0$.

Consider the sum of the first $n$ terms of the series for some large integer $n$, denoted as $S_n$:

$$ S_n = 1 + \frac{1}{1!} + \frac{1}{2!} + \cdots + \frac{1}{n!} $$

Since each term in this sum is positive, $S_n$ is an increasing sequence. Also, $S_n$ is bounded above by $e$, because $e$ is the limit of this sequence as $n$ approaches infinity.

Now, let's choose $n$ such that $b > (n+1)!$. Since $b$ is finite, such an $n$ exists.

If we multiply both sides of $e = \frac{a}{b}$ by $(n+1)!$, we get:

$$ (n+1)! \cdot e = (n+1)! \cdot \frac{a}{b} = a \cdot \frac{(n+1)!}{b} $$

Since $b > (n+1)!$, $\frac{(n+1)!}{b} < 1$, so $a \cdot \frac{(n+1)!}{b}$ is less than $a$.

Now, let's consider $(n+1)! \cdot S_n$:

$$ (n+1)! \cdot S_n = (n+1)! + (n+1)! \cdot \frac{1}{1!} + \cdots + (n+1)! \cdot \frac{1}{n!} $$

Each term of this sum is less than $a$, and there are $a$ terms in total, so $(n+1)! \cdot S_n < a \cdot a = a^2$.

Similarly, $(n+1)! \cdot S_{n+1} > a$.

So, we have $a < (n+1)! \cdot S_n < a^2 < (n+1)! \cdot S_{n+1} < a$. This is a contradiction, because $a$ is an integer.

Therefore, our initial assumption that $e$ is rational must be false, and thus $e$ is irrational.

**3. To prove the triangle inequality:**

\begin{align*}
\lvert x + y \rvert &\leq \lvert x \rvert + \lvert y \rvert, \quad \forall (x,y) \in \mathbb{R} \\
\text{where } x &\in \mathbb{R}, \text{ and} \\
\lvert x \rvert &= \max \{ x, -x \} = \begin{cases} x & \text{if } x \geq 0 \\ -x & \text{if } x \leq 0 \end{cases}
\end{align*}


It follows that $x \leq |x|$ and $-x \leq |x|$.

Now, let's fix $a, b \in \mathbb{R}$:

$$
\begin{aligned}
a + b &\leq |a| + |b| \\
- a -b &\leq |a| + |b| \\
- a - b &= -(a+b)
\end{aligned}
$$

So,

$$
\begin{aligned}
a+b &\leq |a| + |b| \\
-(a+b) &\leq |a| + |b|
\end{aligned}
$$

$$
\max \{ a + b, -(a+b) \} = |a + b | \leq |a| + |b|
$$

**Proof of Sedrakayan's Lemma:**

We start by using the Cauchy-Schwarz Inequality.

The Cauchy-Schwarz Inequality states that for any real numbers $a_1, a_2, ..., a_n$ and $b_1, b_2, ..., b_n$, the following inequality holds:

$$
\left( \sum_{i=1}^{n} a_i b_i \right)^2 \leq \left( \sum_{i=1}^{n} a_i^2 \right) \left( \sum_{i=1}^{n} b_i^2 \right)
$$

Let's use this inequality with $a_i = \sqrt{v_i}$ and $b_i = \frac{u_i}{\sqrt{v_i}}$:

$$
\left( \sum_{i=1}^{n} \sqrt{v_i} \frac{u_i}{\sqrt{v_i}} \right)^2 \leq \left( \sum_{i=1}^{n} (\sqrt{v_i})^2 \right) \left( \sum_{i=1}^{n} \left( \frac{u_i}{\sqrt{v_i}} \right)^2 \right)
$$

This simplifies to:

$$
\left( \sum_{i=1}^{n} u_i \right)^2 \leq \left( \sum_{i=1}^{n} v_i \right) \left( \sum_{i=1}^{n} \frac{u_i^2}{v_i} \right)
$$

Now, divide both sides by $\sum_{i=1}^{n} v_i$:

$$
\frac{\left( \sum_{i=1}^{n} u_i \right)^2}{\sum_{i=1}^{n} v_i} \leq \sum_{i=1}^{n} \frac{u_i^2}{v_i}
$$

This proves Sedrakayan's Lemma.

Given that $(u_i)$ and $(v_i)$ are square roots of even integers, we can express them as $ (u_i = \sqrt{a_i}) $ and $ (v_i = \sqrt{b_i}) $, where $(a_i)$ and $(b_i)$ are even integers.

Expanding both sides of the inequality:

Left-hand side:

$$
\frac{\left(\sum_{i=1}^n u_i\right)^2}{\sum_{i=1}^n v_i} \\
= \frac{\left(\sum_{i=1}^n \sqrt{a_i}\right)^2}{\sum_{i=1}^n \sqrt{b_i}} \\
= \frac{\left(\sqrt{\sum_{i=1}^n a_i}\right)^2}{\sqrt{\sum_{i=1}^n b_i}} \\
= \frac{\sum_{i=1}^n a_i}{\sqrt{\sum_{i=1}^n b_i}}
$$

Right-hand side:

$$
\sum_{i=1}^n \frac{(u_i)^2}{v_i} \\
= \sum_{i=1}^n \frac{a_i}{\sqrt{b_i}}
$$

Prove the inequality $(\frac{\sum_{i=1}^n a_i}{\sqrt{\sum_{i=1}^n b_i}} \leq \sum_{i=1}^n \frac{a_i}{\sqrt{b_i}})$, use the Cauchy-Schwarz Inequality.

For vectors $(\mathbf{x} = (\sqrt{a_1}, \sqrt{a_2}, ..., \sqrt{a_n}))$ and $(\mathbf{y} = (\frac{1}{\sqrt{b_1}}, \frac{1}{\sqrt{b_2}}, ..., \frac{1}{\sqrt{b_n}}))$, the Cauchy-Schwarz Inequality states:

$$
\left(\sum_{i=1}^n x_i y_i\right)^2 \leq \left(\sum_{i=1}^n x_i^2\right) \left(\sum_{i=1}^n y_i^2\right)
$$

Applying this to our case:

$$
\left(\sum_{i=1}^n \sqrt{a_i} \cdot \frac{1}{\sqrt{b_i}}\right)^2 \leq \left(\sum_{i=1}^n (\sqrt{a_i})^2\right) \left(\sum_{i=1}^n \left(\frac{1}{\sqrt{b_i}}\right)^2\right) \\
\left(\sum_{i=1}^n \frac{a_i}{\sqrt{b_i}}\right)^2 \leq \left(\sum_{i=1}^n a_i\right) \left(\sum_{i=1}^n \frac{1}{b_i}\right)
$$

Now, because $(a_i)$ and $(b_i)$ are even integers, $(\frac{1}{b_i})$ is always positive, so $(\sum_{i=1}^n \frac{1}{b_i})$ is also positive. Therefore:

$$
\left(\sum_{i=1}^n \frac{a_i}{\sqrt{b_i}}\right)^2 \leq \left(\sum_{i=1}^n a_i\right) \left(\sum_{i=1}^n \frac{1}{b_i}\right) \\
\sum_{i=1}^n \frac{a_i}{\sqrt{b_i}} \leq \sqrt{\left(\sum_{i=1}^n a_i\right) \left(\sum_{i=1}^n \frac{1}{b_i}\right)}
$$

This implies:

$$
\frac{\sum_{i=1}^n a_i}{\sqrt{\sum_{i=1}^n b_i}} \leq \sum_{i=1}^n \frac{a_i}{\sqrt{b_i}}
$$

Therefore, Sedrakyan's Lemma is proven.
