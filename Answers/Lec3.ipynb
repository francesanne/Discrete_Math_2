#### $\textbf{Bipartite Graph}$

```python
import matplotlib.pyplot as plt
import networkx as nx

seed = 42
# Use seed for reproducibility
G = nx.Graph()

# Add nodes to two separate sets
G.add_nodes_from([0, 1, 2, 3], bipartite=0)  # Nodes in the first set (labeled as 0)
G.add_nodes_from(['a', 'b', 'c'], bipartite=1)  # Nodes in the second set (labeled as 1)

# Add edges between nodes of different sets
edgelist = [(0, 'a'), (0, 'b'), (1, 'a'), (1, 'c'), (2, 'b'), (3, 'c')]
G.add_edges_from(edgelist)

# Use bipartite layout for visualization
pos = nx.bipartite_layout(G, [0, 1, 2, 3])  # Adjust the order of nodes in the first set

nx.draw(G, pos=pos, with_labels=True, node_size=1000, node_color='lightyellow', font_size=12, edge_color='black', width=2)
plt.show()
```

This graph is bipartite, consisting of two sets of nodes. Nodes in the first set are labeled as 0, 1, 2, and 3, while nodes in the second set are labeled as 'a', 'b', and 'c'. Edges are only present between nodes of different sets, forming a bipartite structure. The visualization is done using a bipartite layout, which arranges the nodes in two separate rows, ensuring that nodes from the same set do not overlap.

#### $\textbf{Subgraphs}$

```python
import matplotlib.pyplot as plt
import networkx as nx

seed = 55
# Use seed for reproducibility
G = nx.Graph()

# Add nodes
G.add_nodes_from([6, 7, 8, 9])
# Add edges
edgelist = [(6, 7), (6, 8), (6, 9), (7, 8), (7, 9), (8, 9)]
G.add_edges_from(edgelist)

pos = nx.planar_layout(G)  # Seed for reproducible layout
nx.draw(G, pos=pos, with_labels=True, node_size=1000, node_color='pink', font_size=12, edge_color='black', width=2)
plt.title('Original Graph')
plt.show()

# Generate a subgraph
subgraph_nodes = [6, 7]  # Specify the nodes for the subgraph
subgraph = G.subgraph(subgraph_nodes)

# Draw the subgraph
subgraph_pos = nx.spring_layout(subgraph)  # Use a different layout for the subgraph
nx.draw(subgraph, pos=subgraph_pos, with_labels=True, node_size=1000, node_color='pink', font_size=12, edge_color='black', width=2)
plt.title('Subgraph')
plt.show()
```

The code generates and visualizes a graph with four nodes and six edges. It then extracts a subgraph comprising nodes 6 and 7 from the original graph and displays it separately. The original graph is laid out using a planar layout, while the subgraph layout is determined by a spring layout algorithm, resulting in clear visualization of the selected nodes and edges.

#### $\textbf{The Handshake Theorem}$

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a graph
G = nx.Graph()
G.add_nodes_from([1, 2, 3, 4, 5, 6])
edgelist = [(1, 2), (1, 3), (1, 4), (2, 3), (2, 4), (2, 5), (3, 4), (3, 5), (4, 6)]
G.add_edges_from(edgelist)

# Seed for reproducible layout
seed = 42
pos = nx.spring_layout(G, seed=seed)

# Draw the graph
nx.draw(G, pos=pos, with_labels=True, node_size=1000, node_color='orange', font_size=12, edge_color='black', width=2)
plt.show()
```

The handshake theorem states that the sum of the degrees of all the vertices in a graph is equal to twice the number of edges in the graph.

- Degree of $v_1$ = 3
- Degree of $v_2$ = 4
- Degree of $v_3$ = 4
- Degree of $v_4$ = 4
- Degree of $v_5$ = 2
- Degree of $v_6$ = 1

Total Edges: 9

Total Degree = $3 + 4 + 4 + 4 + 2 + 1 = 18$

Edges * 2 = Total Degree

$9 * 2 = 18$

Since the total degree of the graph is even (18), the graph exists.

#### $\textbf{Travelling in a Graph}$
$\textbf{Walk, Trail, Path, Closed Walk, Circuit, Simple Circuit:}$

```python
import matplotlib.pyplot as plt
import networkx as nx

walk = nx.Graph()
walk.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5), (5, 1), (1, 3)])

trail = nx.Graph()
trail.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5), (5, 3)])

path = nx.Graph()
path.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5)])

closed_walk = nx.Graph()
closed_walk.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])

circuit = nx.Graph()
circuit.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])

simple_circuit = nx.Graph()
simple_circuit.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])


walk_path = [(1, 2), (2, 3), (3, 4), (4, 5), (5, 1), (1, 3)]
trail_path = [(1, 2), (2, 3), (3, 4), (4, 5), (5, 3)]
path_path = [(1, 2), (2, 3), (3, 4), (4, 5)]
closed_walk_path = [(1, 2), (2, 3), (3, 4), (4, 1)]
circuit_path = [(1, 2), (2, 3), (3, 4), (4, 1)]
simple_circuit_path = [(1, 2), (2, 3), (3, 4), (4, 1)]


plt.figure(figsize=(15, 10))

# Walk
plt.subplot(231)
nx.draw(walk, pos=nx.circular_layout(walk), with_labels=True, node_size=500, node_color='lightblue')
nx.draw_networkx_edges(walk, pos=nx.circular_layout(walk), edgelist=walk_path, edge_color='green', width=2)
plt.title('Walk')

# Trail
plt.subplot(232)
nx.draw(trail, pos=nx.planar_layout(trail), with_labels=True, node_size=500, node_color='lightblue')
nx.draw_networkx_edges(trail, pos=nx.planar_layout(trail), edgelist=trail_path, edge_color='red', width=2)
plt.title('Trail')

# Path
plt.subplot(233)
nx.draw(path, pos=nx.spring_layout(path, seed=55), with_labels=True, node_size=500, node_color='lightblue')
nx.draw_networkx_edges(path, pos=nx.spring_layout(path, seed=55), edgelist=path_path, edge_color='blue', width=2)
plt.title('Path')

# Closed Walk
plt.subplot(234)
nx.draw(closed_walk, pos=nx.spring_layout(closed_walk, seed = 56), with_labels=True, node_size=500, node_color='lightblue')
nx.draw_networkx_edges(closed_walk, pos=nx.spring_layout(closed_walk, seed = 56), edgelist=closed_walk_path, edge_color='orange', width=2)
plt.title('Closed Walk')

# Circuit
plt.subplot(235)
nx.draw(circuit, pos=nx.spring_layout(circuit, seed = 57), with_labels=True, node_size=500, node_color='lightblue')
nx.draw_networkx_edges(circuit, pos=nx.spring_layout(circuit, seed = 57), edgelist=circuit_path, edge_color='purple', width=2)
plt.title('Circuit')

# Simple Circuit
plt.subplot(236)
nx.draw(simple_circuit, pos=nx.spring_layout(simple_circuit, seed = 58), with_labels=True, node_size=500, node_color='lightblue')
nx.draw_networkx_edges(simple_circuit, pos=nx.spring_layout(simple_circuit, seed = 58), edgelist=simple_circuit_path, edge_color='brown', width=2)
plt.title('Simple Circuit')

plt.tight_layout()
plt.show()
```

1. **Walk**: A walk is a sequence of vertices and edges where consecutive vertices are connected by edges. In the walk graph, the walk starts from vertex 1 and follows the route: $1 \rightarrow 2 \rightarrow 3 \rightarrow 4 \rightarrow 5 \rightarrow 1 \rightarrow 3$, forming a loop.

2. **Trail**: A trail is a walk where no edge is repeated. In the trail graph, the trail starts from vertex 1 and follows the route: $1 \rightarrow 2 \rightarrow 3 \rightarrow 4 \rightarrow 5 \rightarrow 3$.

3. **Path**: A path is a trail where no vertex is repeated. In the path graph, the path starts from vertex 1 and follows the route: $1 \rightarrow 2 \rightarrow 3 \rightarrow 4 \rightarrow 5$.

4. **Closed Walk**: A closed walk is a walk where the starting and ending vertices are the same. In the closed walk graph, the closed walk starts and ends at vertex 1, following the route: $1 \rightarrow 2 \rightarrow 3 \rightarrow 4 \rightarrow 1$.

5. **Circuit**: A circuit is a closed walk where no edge is repeated, except for the starting and ending vertices. In the circuit graph, the circuit starts and ends at vertex 1, following the route: $1 \rightarrow 2 \rightarrow 3 \rightarrow 4 \rightarrow 1$.

6. **Simple Circuit**: A simple circuit is a circuit where no vertex is repeated, except for the starting and ending vertices. In the simple circuit graph, the simple circuit starts and ends at vertex 1, following the route: $1 \rightarrow 2 \rightarrow 3 \rightarrow 4 \rightarrow 1$.

#### $\textbf{Euler's Circuit}$

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a graph with an Eulerian circuit
G = nx.cycle_graph(6)  # Create a cycle graph with 6 nodes
euler_circuit = list(nx.eulerian_circuit(G))  # Get an Eulerian circuit of the graph

# Plot the graph with the Eulerian circuit
plt.figure(figsize=(6, 4))
pos = nx.circular_layout(G)  # Position nodes in a circle
nx.draw(G, pos, with_labels=True, node_size=500, node_color='violet')
nx.draw_networkx_edges(G, pos, edgelist=euler_circuit, edge_color='black', width=2)
plt.title("Euler's Circuit")
plt.show()
```

$Route: 0 \to 1 \to 2 \to 3 \to 4 \to 5 \to 0$

The graph depicted represents an Eulerian circuit, which is a circuit that traverses every edge in the graph exactly once and returns to the starting vertex. The graph consists of 6 nodes arranged in a circular layout, forming a cycle graph. Each node is labeled with a number from 0 to 5, and there are edges connecting adjacent nodes, creating a cycle. An Eulerian circuit, highlighted in black, is shown traversing all edges of the graph, starting and ending at vertex 0. The Eulerian circuit visits every edge exactly once, demonstrating the property of an Eulerian circuit.

#### $\textbf{Hamiltonian Circuit}$

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a cycle graph with 8 nodes
G = nx.cycle_graph(8)

# Add some extra edges to make it more complex
extra_edges = [(0, 3), (1, 4), (2, 5), (3, 6)]
G.add_edges_from(extra_edges)

# Find a Hamiltonian circuit
hamiltonian_circuit = list(G.nodes()) + [list(G.nodes())[0]]  # Hamiltonian circuit

# Plot the graph with the Hamiltonian circuit
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)  # Position nodes with a spring layout
nx.draw(G, pos, with_labels=True, node_size=500, node_color='lightgray')
nx.draw_networkx_edges(G, pos, edgelist=[(hamiltonian_circuit[i], hamiltonian_circuit[i+1]) for i in range(len(hamiltonian_circuit)-1)], edge_color='red', width=2)
plt.title("Hamiltonian Circuit")
plt.show()
```

$Route: 0 \to 1 \to 2 \to 3 \to 4 \to 5 \to 6 \to 7 \to 0 $

The graph above illustrates a Hamiltonian circuit. A Hamiltonian circuit is a path in a graph that visits each vertex exactly once and returns to the starting vertex. The edges of the Hamiltonian circuit are highlighted in red. The graph consists of a cycle with 8 vertices and some additional edges to increase complexity.

#### $\textbf{Travelling Salesman Problem}$

```python
import matplotlib.pyplot as plt
import networkx as nx

# Use seed for reproducibility
seed = 42

# Create an empty undirected graph
G = nx.Graph()

# Add nodes
G.add_nodes_from(["A", "B", "C", "D"])

# Add edges with weights
edgelist = [("A", "B", 20), ("A", "C", 30), ("A", "D", 40), ("B", "C", 50), ("B", "D", 60), ("C", "D", 70)]
G.add_weighted_edges_from(edgelist)

# Define positions for the nodes
pos = nx.spring_layout(G, seed=seed)

# Draw the graph with edge labels
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_size=500, node_color='pink', font_size=12)
edge_labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_color='red')

# Display the plot
plt.title("Traveling Salesman Problem")
plt.show()
```

Route:
$A \to B \to C \to D \to A = 20 + 50 + 70 + 40 = 180$

This graph represents a Traveling Salesman Problem (TSP), where a salesman must visit a set of cities exactly once and return to the starting city while minimizing the total distance traveled. The cities A, B, C, and D are connected by edges, and each edge is labeled with its weight representing the distance between the cities.

The objective is to find the shortest route that visits each city exactly once and returns to the starting city. In this example, the optimal route is $A \to B \to C \to D \to A$, with a total distance of 180 units. This solution demonstrates how the TSP can be approached using graph theory and optimization techniques.
