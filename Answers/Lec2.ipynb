To show that the expected value $ E[y] = \sum y \cdot P(y) = n \cdot p_i $, where $ p_i $ is the probability of the outcome falling into cell $ i $, we'll proceed step by step.

The expected value $ E[y] $ of a random variable $ y $ is calculated as the sum of the products of each possible outcome $ y $ and its probability $ P(y) $:

$ E[y] = \sum_{\text{all } y} y \cdot P(y) $

Given the joint probability distribution $ P(y) = \frac{n!}{y_1!y_2!y_3! \cdots y_n!} \cdot p_1^{y_1} \cdot p_2^{y_2} \cdot p_3^{y_3} \cdots p_k^{y_k} $, where $ y_1 + y_2 + \cdots + y_k = n $, we can plug it into the formula for $ E[y] $:

$ E[y] = \sum_{\text{all } y} y \cdot \left( \frac{n!}{y_1!y_2!y_3! \cdots y_n!} \cdot p_1^{y_1} \cdot p_2^{y_2} \cdot p_3^{y_3} \cdots p_k^{y_k} \right) $

$ E[y] = \sum_{\text{all } y} y \cdot \left( \frac{n!}{y_1!y_2!y_3! \cdots y_n!} \cdot p_1^{y_1} \cdot p_2^{y_2} \cdot p_3^{y_3} \cdots p_k^{y_k} \right) $

$ E[y] = \sum_{y_1 + y_2 + \cdots + y_k = n} \frac{n!}{y_1!y_2!y_3! \cdots y_n!} \cdot y \cdot p_1^{y_1} \cdot p_2^{y_2} \cdot p_3^{y_3} \cdots p_k^{y_k} $

$ E[y] = \sum_{y_1 + y_2 + \cdots + y_k = n} \frac{n!}{y_1!y_2!y_3! \cdots y_n!} \cdot (y_1 p_1 + y_2 p_2 + \cdots + y_k p_k) $

Now, since $ y_1 + y_2 + \cdots + y_k = n $, the summation reduces to:

$ E[y] = n \cdot (p_1 + p_2 + \cdots + p_k) $

$ E[y] = n \cdot (p_1 + p_2 + \cdots + p_k) $

$ E[y] = n \cdot 1 $

$ E[y] = n $

Thus, we've shown that the expected value $ E[y] = n \cdot p_i $. <br><br>

- $y_5 = 500 - (y_1 + y_2 + y_3 + y_4) = 500 - (100 + 200 + 0 + 200) = 0$ (for ages 65-100)

Substitute these values into the PMF and calculate:

$P(y) = \frac{500!}{100!200!0!200!0!} (0.18)^{100} (0.23)^{200} (0.16)^{0} (0.27)^{200} (0.16)^{0}$

<b>Question 2: </b><br>
To find the expectation of the binomial distribution, $E[y]$, which represents the mean number of successes in $n$ trials, we utilize the formula:

$E[y] = \sum_{k=0}^{n} k \cdot P(y=k)$

Where:
- $k$ is the number of successes,
- $P(y=k)$ is the probability of getting $k$ successes in $n$ trials.

For a binomial distribution with parameters $n$ and $p$, where $n$ is the number of trials and $p$ is the probability of success on a single trial, the probability mass function (PMF) is given by:

$P(y=k) = \binom{n}{k} p^k (1-p)^{n-k}$

Substituting this into the expectation formula:

$E[y] = \sum_{k=0}^{n} k \cdot \binom{n}{k} p^k (1-p)^{n-k}$

$E[y] = \sum_{k=0}^{n} k \cdot \frac{n!}{k!(n-k)!} p^k (1-p)^{n-k}$

$E[y] = \sum_{k=0}^{n} \frac{n!}{(k-1)!(n-k)!} p^k (1-p)^{n-k}$

Now, we can factor out $np$ from the summation:

$E[y] = np \sum_{k=0}^{n} \frac{(n-1)!}{(k-1)!(n-k)!} p^{(k-1)} (1-p)^{n-k}$

$E[y] = np \sum_{k=0}^{n} \binom{n-1}{k-1} p^{(k-1)} (1-p)^{n-k}$

$E[y] = np \sum_{j=0}^{n-1} \binom{n-1}{j} p^{j} (1-p)^{n-1-j}$

Now, observe that the summation is similar to the expansion of $(p + (1-p))^{n-1} = 1^{n-1} = 1$.

Thus, 

$E[y] = np$

This shows that the expectation of the binomial distribution is $np$.
